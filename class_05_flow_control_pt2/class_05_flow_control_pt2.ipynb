## Flow control

**Exercise 1**

Write a program that computes a sum of numbers stored in a list. Solve the task using `while` and `for` loop. Do not use `list.sum()` (only to verify the answer).

**Exercise 2**

Construct a nested `for` loop to print elements of a nested list.

**Exercise 3**

Write a program that takes a list of lists of integers and adds up the elements from all of the nested lists (example below).

    t = [[1, 2], [3], [4, 5, 6]] #result: 21

**Exercise 4**

Write a program that checks sorting order of the list and prints `True` if the list is sorted in ascending order and `False` otherwise. Assume that a list contains elements of the same type. Try to do this task without using `sort` or `sorted`. Remember, that the list may contain more than 3 elements.

    [1, 2, 2] 	# True
    ['b', 'a']	# False

**Exercise 5**

Given a list of roots (stored in a list), compute the value of a polynomial $P(x)$ for $x=2$ or a value given by the user.

$$
P(x) = a_0 + a_1 x + a_2 x^2 + \ldots +  a_n x^n
$$

**Exercise 6**

Find prime numbers lower than 100 using a `for` or a `while` loop.

**Exercise 7**

Given the values of coordinates in the lists below, compute 3d distances between the adjusting points:
    
    X = [10, 1, 21, 4.8, 23, 5.6, 98.4] # X coordinates
    Y = [2, 1.5, 4.8, 12.8, 17, 16.8, 98.4] # Y coordinates
    Z = [0, 11, 13, 14.8, 3, 5, 98.4] # Z coordinates

**Exercise 8**

Write a program that takes a string and prints the letters in decreasing order of frequency. Find text samples from several different languages and see how letter frequency varies between languages (exercise adapted from *"Think Python* by A. Downey).

**Exercise 9**

Using list comprehensions, create a list of odd values from 0 to 20.

**Exercise 10**

Create a dictionary containing products and their prices. Create a list with names of some of the products (shopping list). Compute the cost of the products on the shopping list.
