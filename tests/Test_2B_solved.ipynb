## Question 1
Write a function that takes two arguments: a `ndarray` array and a parameter `a`. The function should modify the array as follows:
1. All values in the array divisible by 5 should be switched to -3.
2. Add a new column that will contain the result of multiplying the first column by `a`.
3. Return the array without the last row (remove the last row).
4. Call the function. Print the number of columns of the returned result.

**`first` in the above task means "with index = 0"**<br>
**Hint: You might need to use reshape to get the matching number of dimensions**

**Example**:
input:<br>
`[[ 0,  3,  6,  9],
   [12, 15, 18, 21],
   [24, 27, 30, 33],
   [36, 39, 42, 45],
   [48, 51, 54, 57]]`
   
output:<br>
`[[ -3   3   6   9  -9]
 [ 12  -3  18  21  36]
 [ 24  27  -3  33  72]
 [ 36  39  42  -3 108]]`


```python
def modify_array(ar, a):
    ar[ar % 5 == 0] = -3
    new_col = ar[:, 0] * a
    #print(new_col.shape, ar.shape)
    #print(last_row)
    ar = np.hstack((ar, new_col.reshape(-1, 1)))
    return ar[:-1, :]

x = np.array([[ 0,  3,  6,  9],
   [12, 15, 18, 21],
   [24, 27, 30, 33],
   [36, 39, 42, 45],
   [48, 51, 54, 57]])
x_modified = modify_array(x, 3)
print(x_modified, x_modified.shape[1])
```

## Question 2
Write a function called `remove_element` that takes two arguments: `i` (type: `int`) and an argument of type: `list`, removes the element with index `i` from the list and returns the modified list. Use `try\except` block to avoid situation when the index is out of range. Specify the Exception subclass. Add a message for the user reflecting the situation.

**Hint: You can use the built-in list methods to remove element with a given index.**<br>

**Example:**<br>
`example_list = [0, 1, 2, 3, 4, 5]
print(remove_element(example_list, 2)) -> output: [0, 1, 3, 4, 5]
print(remove_element(example_list, 2)) -> output: Index 8 not in list with length 5.`


```python
def remove_element(l_example, i):
    try:
        l_example.pop(i)
    except IndexError:
        print(f"Index {i} not in list with length {len(l_example)}.")
    else:
        return l_example
    
example_list = [0, 1, 2, 3, 4, 5]
print(remove_element(example_list, 2))
print(remove_element(example_list, 8))
```

## Question 3
Create a class called `RightTriangle` that would create a right triangle objects. 
1. The class should have two attributes: lengths of sides adjacent to the right angle (`a` and `b` in the picture), and the method initializing the triangle with given sides.
2. The class should have two additional methods for computing the area and the hypotenuse (the third side).
3. Create an object of the `RightTriangle` class with `a=3` and `b=4`, compute the area and the hypotenuse.



**Example**<br>
`For right triangle with sides 3 and 4, hypotenuse = 5.000 and area = 6.000.`

```python
from math import sqrt

class RightTriangle:
    """
    Right triangle with two given sides
    """
    
    def __init__(self, a, b):
        # we should check that a& b are > 0
        self.a = a
        self.b = b
        
    def compute_hypotenuse(self):
        return sqrt(self.a**2 + self.b**2)
    
    def compute_area(self):
        return self.a * self.b / 2
    
t = RightTriangle(3, 4)
print(f"For right triangle with sides {t.a} and {t.b}, hypotenuse = {t.compute_hypotenuse():.3f} and area = {t.compute_area():.3f}.")
```

## Question 4
The file `2B.txt` contains three lines of data, values in each line are separated with a `-`. The first column contains the names of CORS stations, the second contains their latitudes and the third -- longitudes (both given in degrees). However, there was a mistake in saving data! Each latitude in a file is 2 degrees too low, and each longitude is 15 degrees to high.
1. Read the data in file.
2. Correct the coordinates.
3. Save data to a file, but this time, the first row should contain names, the second - latitudes and third - longitudes. Separate values in a row with a space.

*Hint: remember that depending on the file reading method you use, latitude and longitude data may be read as str*

**Example**:
input:<br>
`POTS00DEU-50.3792972222222-28.0660916666667
 M0SE00IT-39.8931138888889-27.4931138888889
NICO00CYP-33.1409805555556-48.3964333333333
ONSA00SWE-55.3952972222222-26.9255138888889`

output:<br>
`POTS00DEU M0SE00IT NICO00CYP ONSA00SWE
52.379297 41.893114 35.140981 57.395297
13.066092 12.493114 33.396433 11.925514`

```python
names = []
lat = []
lon = []

with open("2B.txt", "r") as f:
    for line in f:
        line = line.strip().split("-")
        names.append(line[0])
        lat.append(float(line[1])+2)
        lon.append(float(line[2])-15)

lat = [f"{l:.6f}" for l in lat]
lon = [f"{l:.6f}" for l in lon]
# save the data:
with open("2B_new.txt", "w") as new_file:
    new_file.write(" ".join(names) + "\n")
    new_file.write(" ".join(lat) + "\n")
    new_file.write(" ".join(lon))
```

## Question 5


1. Find the value of $\sqrt{10}$ using your own implementation of the Newton-Raphson root-finding method (using scipy method won't give you points for this part of the task).<br>
In Newton-Raphson algorithm the next approximation of a root of a function $f(x)$, $x_{i+1}$ is defined as:
$$ 
x_{i+1} = x_{i} - \frac{f(x_{i})}{f'(x_{i})}
$$
where $f'(x_i)$ is a derivative of function $f(x)$ at point $x_{i}$.
<br>
Stop computation when:
        \begin{equation*}
            \lvert x_{i+1} - x_i \rvert \leq 10^{-4},
        \end{equation*}
where $x_i$ is the $i$-th approximation of the solution.<br>
Save the intermediate solutions.<br><br>

2. Create a figure with two subplots in two rows. The first subplot should contain a line representing $f(x)=x^2-10$ for $x \in [-5, 5]$. The second row should contain a plot of $f(x)$ for $x \in [0, 5]$ and a scatter plot of the approximations from part 1. Change the color of the line to red.<br> 
Add a title, axis description and legend. Set the figure size to (6, 14).


```python
import matplotlib.pyplot as plt
import math as m


def fx(x):
    return x**2 - 10

def fprime(x):
    return 2 * x

def next_approx(f, fprime, xi):
    return xi - f(xi) / fprime(xi)

# define a list with an arbitrary initial value
x = [1]
k = 0
tolerance = 0.0001

# limit iterations
while k < 100:
    k += 1
    x.append(next_approx(fx, fprime, x[-1]))
    if abs(x[-2] - x[-1]) <= tolerance:
        solution = x[-1]
        break
print(solution, m.sqrt(10))
```

```python
x_range = np.linspace(-5, 5)
f = fx(x_range)
# compute values for approximations
x_ap = np.array(x)
f_ap = fx(x_ap)

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6, 14))
ax[0].plot(x_range, f, label="f(x)")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[0].legend()

ax[1].plot(x_range, f,'r', label="f(x)")
ax[1].scatter(x_ap, f_ap, label="approx")
ax[1].set_xlim([0, 5])
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
ax[1].legend()
fig.suptitle("Approximation of f(x)")
plt.show()
```
